```python
import pandas as pd
```

```python
fly_genome = pd.read_csv('gff.tsv', sep='\t')
fly_genome
```

```python
# Создадим функцию, которая: 
# 1. Будет делить серию на листы через split, 
# 2. Будет искать значения, где есть Parent
# 3. Будет удалять значение "Parent="

def find_parent_value(x):
    x = x.split(';')
    for i in x:
        if 'Parent' in i:
            i = i.replace('Parent=', '')
            i = i.replace(',', '')
            return i
```

```python
# Применяем функцию ко всем значениям колонки attributes и записываем значения в новую колонку parent_value
fly_genome['parent_value'] = fly_genome.attributes.apply(find_parent_value)
fly_genome
```

```python
# Посчитаем количество записей по каждому значению колонки parent_value
fly_genome.parent_value.value_counts()
```
